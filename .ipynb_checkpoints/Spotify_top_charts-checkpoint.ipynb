# Spotify_Top_Charts

Download csv of choice at https://spotifycharts.com/regional/

## Set up

```python
import spotipy
import spotipy.util as util
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re #regular expressions
import os
from spotipy.oauth2 import SpotifyClientCredentials
%matplotlib inline
```

We use this function to flatten results from spotify output call. Useful when using apply method in pandas in conjunction with spotify methods

```python
def flatten(x):
    return x[0]
```

Let's import the dataset

```python
dataset = "regional-global-daily-latest.csv"
df = pd.read_csv(dataset,header= 1).set_index("Position")
```

Split URL to get ID, in order to get song attributes

```python
df["ID"] = df["URL"].str.split("track/",expand=True)[1]
```

```python
df.head()
```

## Adding Audio Features

```python
client_id = ""
client_secret = ""
redirect_uri = ''
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id,
                                                      client_secret)
token = client_credentials_manager.get_access_token()
spotify = spotipy.Spotify(auth=token)
```

```python
df = df.join(df["ID"].apply(spotify.audio_features).
              apply(flatten).
              apply(pd.Series))
```

## Analysis

```python
import seaborn as sns
import matplotlib.ticker as ticker
from sklearn import datasets, linear_model
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import TSNE
```

```python
specific = ["energy", "liveness", "tempo", "valence", "loudness", "speechiness", "acousticness", "danceability", "instrumentalness"]
```

```python
for i in specific:
    df[i] = pd.to_numeric(df[i])
```

```python
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

```python
text1 = df["Artist"] + " - " + df["Track Name"]
text2 = text1.values
```

```python
X = df[specific].values
y = df["danceability"].values
```

```python
min_max_scaler = MinMaxScaler()
X = min_max_scaler.fit_transform(X)
```

```python
pca = PCA(n_components=3)
pca.fit(X)

X = pca.transform(X)
```

```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
```

```python
trace = go.Scatter3d(
    x=X[:,0],
    y=X[:,1],
    z=X[:,2],
    text=text2,
    mode="markers",
    marker=dict(
        size=12,
        color=y
    )
)
```

```python
fig = go.Figure(data=[trace])
```

```python
py.iplot(fig, filename="test-graph")
```
