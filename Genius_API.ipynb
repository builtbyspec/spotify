```python
tknzr = TweetTokenizer()
def perc_unique(name):
    url = genius_songs[name][1]
    lyrics = spotfuncs.get_lyrics(url)
    tokens = tknzr.tokenize(lyrics)
    try:
        return (len(pd.Series(tokens).unique())/len(pd.Series(tokens)))
    except:
        return np.nan

def get_name(k):
    name = genius_songs[k][0]
    return name
```
